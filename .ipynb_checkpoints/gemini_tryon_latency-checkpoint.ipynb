# Gemini Try-On Latency Benchmark\n\nThis notebook benchmarks end-to-end latency for try-on style image generation using Gemini `generateContent` across model variants and prompt variants.\n\n- Measures request wall time (`latency_s`)\n- Captures HTTP status / errors\n- Reports whether inline image output was returned\n- Optionally saves one generated sample to disk\n

```python
from __future__ import annotations\n\nimport base64\nimport json\nimport mimetypes\nimport os\nimport re\nimport time\nfrom pathlib import Path\nfrom typing import Any, Dict, List, Tuple\nfrom urllib import error, parse, request\n
```

```python
def load_google_api_key(creds_path: str | Path = "creds.txt") -> str:\n    for k in ("GOOGLE_API_KEY", "GEMINI_API_KEY", "LLM_API_KEY"):\n        v = os.getenv(k)\n        if v and v.strip():\n            return v.strip()\n\n    p = Path(creds_path)\n    if not p.exists():\n        raise FileNotFoundError(f"Missing creds file: {p}")\n    txt = p.read_text(encoding="utf-8", errors="replace")\n    for line in txt.splitlines():\n        raw = line.strip()\n        if not raw or raw.startswith("#"):\n            continue\n        if "google" in raw.lower() or "gemini" in raw.lower():\n            m = re.search(r"=\\s*['\"]?([A-Za-z0-9_\\-]+)['\"]?", raw)\n            if m:\n                return m.group(1).strip()\n    raise ValueError("Could not parse Google API key from creds")\n\n\ndef fetch_image_inline(image_url: str, timeout_s: int = 60) -> Dict[str, Any]:\n    req = request.Request(image_url, method="GET")\n    with request.urlopen(req, timeout=timeout_s) as resp:\n        data = resp.read()\n        ctype = (\n            resp.headers.get_content_type()\n            if hasattr(resp.headers, "get_content_type")\n            else resp.headers.get("Content-Type")\n        )\n    mime = (ctype.split(";")[0].strip() if ctype else None) or mimetypes.guess_type(image_url)[0] or "image/jpeg"\n    return {"inline_data": {"mime_type": mime, "data": base64.b64encode(data).decode("ascii")}}\n\n\ndef extract_inline_image(payload: Dict[str, Any]) -> Tuple[bytes | None, str | None]:\n    for cand in payload.get("candidates", []) or []:\n        parts = ((cand.get("content") or {}).get("parts") or [])\n        for p in parts:\n            inline = p.get("inlineData") or p.get("inline_data")\n            if isinstance(inline, dict) and inline.get("data"):\n                mime = str(inline.get("mimeType") or inline.get("mime_type") or "image/png")\n                return base64.b64decode(inline["data"]), mime\n    return None, None\n
```

```python
# --- Config ---\nAPI_BASE = "https://generativelanguage.googleapis.com"\nCREDS_PATH = "creds.txt"\n\n# Replace with your own product/person image URLs if needed\nPRODUCT_IMAGE_URL = "https://weariq-storage.s3.ap-south-1.amazonaws.com/catlog_test/1164.jpg"\nPERSON_IMAGE_URL = "https://weariq-storage.s3.ap-south-1.amazonaws.com/user-test/sample.jpg"\n\nMODELS = [\n    "gemini-3.1-flash-image-preview",\n    "gemini-3-pro-image-preview",\n    "gemini-2.5-flash-image",\n    "gemini-2.5-flash",\n]\n\nPROMPTS = [\n    "Photorealistic virtual try-on. Preserve person identity, pose, body, and face. Transfer only garment from product image with natural fit and fabric drape.",\n    "Photorealistic virtual try-on. Preserve person identity, pose, body, and face. Transfer only garment from product image with natural fit and fabric drape. Keep lighting and background from person image unchanged. Return one final image only.",\n]\n\nUSE_RESPONSE_MODALITIES = [True, False]\nTIMEOUT_SECONDS = 180\n\napi_key = load_google_api_key(CREDS_PATH)\nprint("Key loaded. Model count:", len(MODELS))
```

```python
# Prepare image inline payload once so benchmark reflects model latency mostly\nproduct_inline = fetch_image_inline(PRODUCT_IMAGE_URL, timeout_s=60)\nperson_inline = fetch_image_inline(PERSON_IMAGE_URL, timeout_s=60)\nprint("Prepared inline images")
```

```python
def run_once(model: str, prompt: str, use_response_modalities: bool) -> Dict[str, Any]:\n    url = f"{API_BASE}/v1beta/models/{model}:generateContent?" + parse.urlencode({"key": api_key})\n    body: Dict[str, Any] = {\n        "contents": [\n            {\n                "role": "user",\n                "parts": [\n                    {"text": prompt},\n                    person_inline,\n                    product_inline,\n                ],\n            }\n        ]\n    }\n    if use_response_modalities:\n        body["generationConfig"] = {"responseModalities": ["TEXT", "IMAGE"]}\n\n    payload = json.dumps(body).encode("utf-8")\n    req = request.Request(\n        url=url,\n        data=payload,\n        headers={"Content-Type": "application/json"},\n        method="POST",\n    )\n\n    t0 = time.perf_counter()\n    try:\n        with request.urlopen(req, timeout=TIMEOUT_SECONDS) as resp:\n            raw = resp.read().decode("utf-8")\n            dt = time.perf_counter() - t0\n            out = json.loads(raw) if raw else {}\n            img_bytes, img_mime = extract_inline_image(out)\n            return {\n                "ok": True,\n                "status": resp.status,\n                "latency_s": round(dt, 3),\n                "has_image": img_bytes is not None,\n                "image_mime": img_mime,\n                "payload": out,\n                "error": None,\n            }\n    except error.HTTPError as exc:\n        dt = time.perf_counter() - t0\n        detail = exc.read().decode("utf-8", errors="replace")\n        return {\n            "ok": False,\n            "status": exc.code,\n            "latency_s": round(dt, 3),\n            "has_image": False,\n            "image_mime": None,\n            "payload": None,\n            "error": detail[:2000],\n        }\n
```

```python
results: List[Dict[str, Any]] = []\n\nfor model in MODELS:\n    for i, prompt in enumerate(PROMPTS, start=1):\n        for rm in USE_RESPONSE_MODALITIES:\n            r = run_once(model, prompt, rm)\n            row = {\n                "model": model,\n                "prompt_variant": i,\n                "use_response_modalities": rm,\n                "ok": r["ok"],\n                "status": r["status"],\n                "latency_s": r["latency_s"],\n                "has_image": r["has_image"],\n                "error": (r["error"][:120] + "...") if r["error"] else None,\n            }\n            results.append({**row, "_raw": r})\n            print(row)\n\nprint("Done. Total runs:", len(results))
```

```python
# Pretty summary without pandas dependency\nsummary = []\nfor r in results:\n    summary.append({k: r[k] for k in ("model", "prompt_variant", "use_response_modalities", "ok", "status", "latency_s", "has_image", "error")})\nsummary
```

```python
# Save first successful image output (if any)\nout_dir = Path("./tmp_tryon_outputs")\nout_dir.mkdir(exist_ok=True)\n\nsaved = None\nfor r in results:\n    raw = r["_raw"]\n    if raw["ok"] and raw["has_image"] and raw["payload"]:\n        img_bytes, img_mime = extract_inline_image(raw["payload"])\n        if img_bytes:\n            ext = ".png" if (img_mime and "png" in img_mime.lower()) else ".jpg"\n            p = out_dir / f"sample_{r['model'].replace('/', '_')}_p{r['prompt_variant']}_{'rm' if r['use_response_modalities'] else 'nrm'}{ext}"\n            p.write_bytes(img_bytes)\n            saved = str(p)\n            break\n\nprint("saved:", saved)
```
